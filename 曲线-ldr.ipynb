7, 星形线

$ x=a \cos{\theta}^3, y=a \sin{\theta}^3$

```python
pi = np.pi
a = 1
plt.figure(figsize=(3,3))
theta = np.linspace(0,2*pi,50)
x = np.cos(theta)**3
y = np.sin(theta)**3

plt.plot(x,y,':o')
plt.axis('equal');
```

摆线: $x=a(\theta-\sin{\theta}), y=a(1-\cos{\theta})$

```python
pi = np.pi
a = 1
plt.figure(figsize=(3,3))
theta = np.linspace(-pi/2,3*pi,50)
x = a*(theta - np.sin(theta))
y = a*(1 - np.cos(theta))
plt.plot(x,y,':o')
plt.axis('equal');
```

```python
plt.polar(x,y,'o:')
plt.axis('off')
```

心形线: $\rho=a(1-\cos{\theta})$

```python
a = 1
plt.figure(figsize=(3,3))
theta = np.linspace(0,2*pi,20)
rho = a*(1 - np.cos(theta))
plt.polar(theta,rho,':o')
plt.axis('off')
```

阿基米德螺线: $\rho=a\theta$

```python
a = 1
plt.figure(figsize=(3,3))
theta = np.linspace(0,2*pi,50)
rho = a*theta
plt.polar(theta,rho,':o')
```

对数螺线$\rho=e^{a\theta}$

```python
a = 1
plt.figure(figsize=(5,5))
theta = np.linspace(0,3*pi,20)
rho = np.exp(a*theta)
plt.polar(theta,rho,'o')
```

双曲螺线: $\rho\theta=a$

```python
a = 1
plt.figure(figsize=(5,5))

theta = np.linspace(pi/12,2*pi,20)
rho = a/theta
plt.polar(theta,rho,'o')
```

伯努利双纽线$\rho^2=a^2\sin{2\theta}$

```python
a = 1
plt.figure(figsize=(5,5))

theta = np.linspace(0,pi/2,100)
rho = a* np.sqrt( np.sin(2*theta) )

theta2 = np.linspace(pi,3*pi/2,100)
rho2 = a* np.sqrt( np.sin(2*theta) )

plt.polar(theta, rho,'+:')
plt.polar(theta2, rho2,'+:')
```
